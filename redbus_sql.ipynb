```python
a=",".join(f"{i} {j}"
for i,j in zip(df.columns,df.dtypes)).replace("object","text").replace("float64","float").replace("int64","int")
```

```python
f"create table statename (id int auto_increment primary key,{a})"
```

```python
myconnection.cursor().execute(f"create table redbus.statename (id int auto_increment primary key,{a})")
```

```python
sql="insert into redbus.statename(busroute,links,busname,bustype,departure,arrival,traveltime,busrating,busfare,busseats) values"
for i in range(len(df)):
    myconnection.cursor().execute(f"{sql} {tuple(df.iloc[i])}")
    myconnection.commit()
```
